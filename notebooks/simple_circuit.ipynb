```python
import perceval as pcvl
import perceval.components as comp
```

```python
c = pcvl.Circuit(2)
c.add(0, comp.BS())
pcvl.pdisplay(c)
```

```python
pcvl.pdisplay(c.U)
c.U.is_unitary()
```

```python
input_state = pcvl.BasicState("|0,1>")
input_state.n
```

```python
print(input_state)
```

```python
backend = pcvl.BackendFactory.get_backend("SLOS")
backend.set_circuit(c)
```

```python
input_state = pcvl.BasicState([1, 0])
output_state = pcvl.BasicState([0, 1])

backend.set_input_state(input_state)
prob_amplitude = backend.prob_amplitude(output_state)
probability = abs(prob_amplitude)**2

print(f"Probabilidade de {input_state} -> {output_state}: {probability:.2f}")
```
